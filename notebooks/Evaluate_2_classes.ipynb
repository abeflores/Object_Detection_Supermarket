## Evaluation of trained model (YOLO v5) with 2 classes


In this notebook, the model is evaluated with three different inference configuration. By varying the confidence threshold and the intersection over union threshold, better results can be achieved both in the metrics and the quality

```python
import cv2
import matplotlib.pyplot as plt
import os

detect_script = '/src/yolov5/detect.py'
validation_script = '/src/yolov5/val.py'
data_test = '/src/datasets/SKU-110K-test/SKU-110K-test.yaml'
weights_file = '/src/yolov5/weights/model.pt'
```

## Confidence Threshold = 0.001, IOU Threshold = 0.6 (Default)

|    Class   |  Images  |   Labels      |    P    |      R  |   mAP@.5   |    mAP@  |
| --- | --- | --- | --- | --- | --- | --- 
|      all   |     139  |    21253      | 0.865   |   0.612 |     0.694  |     0.42 |
|   object   |     139  |    20264      | 0.899   |   0.861 |     0.901  |    0.579 |
|no_object   |     139  |      989      | 0.831   |   0.363 |     0.487  |    0.262 |

```python
!python $validation_script --data $data_test --weights $weights_file --conf-thres 0.001 --iou-thres 0.6 --task test --workers 0
```

### F1 Curve

```python
results_path = '/src/yolov5/runs/val/exp9'
fig = plt.figure(figsize = (15,15))
img = cv2.imread(os.path.join(results_path, 'F1_curve.png'))
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

### PR Curve

```python
fig = plt.figure(figsize = (15,15))
img = cv2.imread(os.path.join(results_path, 'PR_curve.png'))
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

### Confusion Matrix

```python
fig = plt.figure(figsize = (15,15))
img = cv2.imread(os.path.join(results_path, 'confusion_matrix.png'))
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```
